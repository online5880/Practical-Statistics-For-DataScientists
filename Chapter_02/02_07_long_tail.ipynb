```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
```

```python

sp500_px = pd.read_csv('../제공자료/data/sp500_data.csv.gz')

nflx = sp500_px.NFLX
nflx = np.diff(np.log(nflx[nflx>0]))

fig, ax = plt.subplots(figsize=(4, 4))
stats.probplot(nflx, plot=ax)

plt.tight_layout()
plt.show()
```
